```python
# Open a dataset
# Use a library called "Pandas

import pandas as pd
```

```python
# open the dataset
df = pd.read_csv('./yt_data_sb.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# prompt: remove duplicate records

df.drop_duplicates(inplace=True)
df
```

```python
# handle missing values
```

```python
# correcting data types
df.dtypes
```

```python
df.isnull().sum()
```

```python
# prompt: remove missing values so there isn't any nulls

# remove rows with missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# prompt: see if there is a correlation between dislikes and whether or not video comments are disabled and visualize the data.

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'comments_disabled' is a boolean column (True/False) and 'dislikes' is numeric
# If not, adjust the code accordingly

# Calculate the correlation
correlation = df['dislikes'].corr(df['comments disabled'].astype(int))  # Convert boolean to int for correlation

print(f"Correlation between dislikes and comments disabled: {correlation}")

# Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x='comments disabled', y='dislikes', data=df)
plt.title('Dislike Count vs. Comments Disabled')
plt.xlabel('Comments Disabled')
plt.ylabel('Dislikes')
plt.show()


# Alternative Visualization (scatter plot, might be less informative for boolean x-axis)
plt.figure(figsize=(8,6))
sns.scatterplot(x='comments disabled', y='dislikes', data=df)
plt.title('Dislike Count vs. Comments Disabled')
plt.xlabel('Comments Disabled')
plt.ylabel('Dislikes')
plt.show()
```

```python
# prompt: see if there is a correlation between likes and views and visualize the data.



# Assuming 'likes' and 'views' are columns in your DataFrame 'df'
# Calculate the correlation
correlation = df['likes'].corr(df['views'])
print(f"Correlation between likes and views: {correlation}")

# Visualization using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='likes', data=df)
plt.title('Likes vs. Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

# Visualization using a heatmap (for a more compact representation)
plt.figure(figsize=(6, 4))
sns.heatmap(df[['likes', 'views']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Likes vs. Views')
plt.show()
```

```python
# prompt: see if there is a correlation between views and comments and visualize the data.


# Assuming 'views' and 'comment count' are columns in your DataFrame 'df'
# Calculate the correlation
correlation = df['views'].corr(df['comment count'])
print(f"Correlation between views and comment count: {correlation}")

# Visualization using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='comment count', data=df)
plt.title('Comment Count vs. Views')
plt.xlabel('Views')
plt.ylabel('Comment Count')
plt.show()

# Visualization using a heatmap (for a more compact representation)
plt.figure(figsize=(6, 4))
sns.heatmap(df[['views', 'comment count']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Views vs. Comment Count')
plt.show()
```

```python
# prompt: which category id trends the most



# Assuming 'categoryId' is the column representing video categories
category_counts = df['category id'].value_counts()

# Find the category ID with the most trends
most_frequent_category = category_counts.index[0]
print(f"The category ID that trends the most is: {most_frequent_category}")


# Visualization
plt.figure(figsize=(12, 6))
sns.countplot(x='category id', data=df, order=category_counts.index)
plt.title('Trend of Video Categories')
plt.xlabel('Category ID')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# prompt: explore the relationship between publish time and video popularity



# Assuming 'publish_time' and 'views' are columns in your DataFrame 'df'

# Convert 'publish_time' to datetime objects if it's not already
df['publish time'] = pd.to_datetime(df['publish time'])

# Extract relevant time features (e.g., day of the week, hour of the day)
df['publish_day'] = df['publish time'].dt.dayofweek  # Monday=0, Sunday=6
df['publish_hour'] = df['publish time'].dt.hour

# Explore the relationship between publish time and views
# Example: Day of the week vs. views
plt.figure(figsize=(10, 6))
sns.boxplot(x='publish_day', y='views', data=df)
plt.title('Views vs. Publish Day of the Week')
plt.xlabel('Day of the Week (0=Monday, 6=Sunday)')
plt.ylabel('Views')
plt.show()

# Example: Hour of the day vs. views
plt.figure(figsize=(12, 6))
sns.boxplot(x='publish_hour', y='views', data=df)
plt.title('Views vs. Publish Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Views')
plt.show()

# Calculate the correlation between publish time features and views
print(df['publish_day'].corr(df['views']))
print(df['publish_hour'].corr(df['views']))

# You can explore other time features (month, year, etc.) and use different visualization techniques
# (scatter plots, violin plots) to further analyze the relationship.
```

```python
# prompt: Analyzing the distribution of likes, dislikes, and comments

# Assuming 'likes', 'dislikes', and 'comment_count' are columns in your DataFrame 'df'

# Distribution of Likes
plt.figure(figsize=(10, 6))
sns.histplot(df['likes'], kde=True)  # kde=True adds a kernel density estimate
plt.title('Distribution of Likes')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.show()

# Distribution of Dislikes
plt.figure(figsize=(10, 6))
sns.histplot(df['dislikes'], kde=True)
plt.title('Distribution of Dislikes')
plt.xlabel('Number of Dislikes')
plt.ylabel('Frequency')
plt.show()

# Distribution of Comment Count
plt.figure(figsize=(10, 6))
sns.histplot(df['comment count'], kde=True)
plt.title('Distribution of Comment Count')
plt.xlabel('Number of Comments')
plt.ylabel('Frequency')
plt.show()

# Combined visualization (box plots)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['likes', 'dislikes', 'comment count']])
plt.title('Distribution of Likes, Dislikes, and Comment Count')
plt.ylabel('Count')
plt.show()

# Calculate and print some descriptive statistics
print("Likes Statistics:")
print(df['likes'].describe())

print("\nDislikes Statistics:")
print(df['dislikes'].describe())

print("\nComment Count Statistics:")
print(df['comment count'].describe())
```
